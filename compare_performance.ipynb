```python
import matplotlib.image as mpimg
from matplotlib import rcParams
import matplotlib.pyplot as plt
```

```python
model_id = '1638198758'

los_img = mpimg.imread(f'./outputs/{model_id}/loss.png')
acc_img = mpimg.imread(f'./outputs/{model_id}/accuracy.png')

rcParams['figure.figsize'] = 11, 8

_, ax = plt.subplots(1, 2)

ax[0].imshow(los_img)
ax[0].axis('off')

ax[1].imshow(acc_img)
ax[1].axis('off')
```
